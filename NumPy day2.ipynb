```python
n = 100
draws = np.random.randint(0,2, size = n)
steps = np.where(draws > 0, 1, -1)
walk = steps.cumsum()
walk
```

```python
plt.plot(np.arange(0, n, 1), walk)
plt.plot(np.arange(0, n, 1), np.zeros(n))
```

```python
np.abs(walk)
```

```python
np.sum(walk) / 100
```

```python
(np.abs(walk) >= 10).argmax()
```
